```python
import datetime

class bank:
    bname = "DIVINE BANK"
    branch = 'YARAM NAGAR'
    ifsc = 'DIVI0342'
    toll_no =8075000354
    roi = 8
    no_cust = 0
    cust_details = {}
    cust_transactions = {}

    def __init__(self, name, age, email, phone, address, adhaar, pancard, dob, gender, pin, bal):
        self.name = name
        self.age = age
        self.email = email
        self.phone = phone
        self.address = address
        self.adhaar = adhaar
        self.pancard = pancard
        self.dob = dob
        self.gender = gender
        self.pin = pin
        self.bal = bal
        self.no_cust_inc()
        self.account_no = 350 + self.no_cust
        self.concat_cust_details(self.account_no, self)

    @classmethod
    def no_cust_inc(cls):
        cls.no_cust += 1

    @classmethod
    def concat_cust_details(cls, acc_no, obj):
        cls.cust_details[acc_no] = obj

    @classmethod
    def change_ROI(cls):
        cls.roi = int(input('Enter the ROI:'))
        print('ROI is changed')

    @classmethod
    def Check_Bal(cls):
        acc_no = int(input("Enter the Account NO:"))
        pin = int(input("Enter the account pin no: "))
        for i in cls.cust_details:
            res = cls.cust_details[i]
            if res.account_no == acc_no and res.pin == pin:
                print(res.account_no, "balance is:", res.bal)
                return
        else:
            print("Invalid acc_no ")

    @classmethod
    def concat_cust_tran(cls, acc_no, data):
        if acc_no in cls.cust_transactions:
            cls.cust_transactions[acc_no] += [data]
            print(acc_no, data)
        else:
            cls.cust_transactions[acc_no] = [data]
            print(acc_no, data)

    def deposit(self, count=0):
        if count == 3:
            return
        amt = int(input("Enter the Deposit amount: "))
        if amt >= 50:
            self.bal += amt
            self.concat_cust_details(self.account_no, {'date': datetime.datetime.now(), 'type': 'deposit', 'amount': amt, 'bal': self.bal})
            self.success()
            return
        else:
            print('Amount should be a minimum of 50 rupees')
            self.failed()
            return self.deposit(count + 1)

    def withdraw(self, count=0):
        if count == 3:
            return
        amt = int(input('Enter the withdraw amount:'))
        if amt >= 100 and self.bal >= amt:
            self.bal -= amt
            self.concat_cust_tran(self.account_no, {'date': datetime.datetime.now(), 'type': 'withdraw', 'amount': amt, 'bal': self.bal})
            self.success()
            return
        else:
            print('Amount should be a minimum of 100 rupees')
            self.failed()
            return self.withdraw(count + 1)

    def mini_statement(self):
        print('Date'.center(25), 'Type'.center(15), 'Amount'.center(15), "Balance".center(15), sep='|')
        print(80 * '-')
        for i in bank.cust_transactions[self.account_no]:
            print(str(i['date']).center(25), str(i['type']).center(15), str(i['amount']).center(15), str(i['bal']).center(15), sep='|')
            print(80 * '-')

    def check_bal(self):
        print("Your current balance:", self.bal)

    def pin_change(self):
        old = int(input('Enter the old pin:'))
        if self.pin == old:
            self.pin = int(input('Enter the new pin:'))
            print('Pin changed')
            self.success()
        else:
            self.failed()

    def cust_update(self):
        n = int(input("Choose the option:\n1.Name\n2.Phone\n3.Email\n4.Address\n5.Exit:"))
        match n:
            case 1:
                self.name = input("Enter the Name:")
            case 2:
                self.phone = int(input("Enter the phone"))
            case 3:
                self.email = input("Enter the Email")
            case 4:
                self.address = input("Enter the address")
            case _:
                return

    def account_details(self):
        print("Acc no:", self.account_no)
        print("Acc Holder:", self.name)
        print("Age:", self.age)
        print("Gmail:", self.email)
        print("Mobile:", self.phone)
        print("Balance:", self.bal)
        print("IFSC code:", self.ifsc)
        print("PAN:", self.pancard)
        print("Aadhaar:", self.adhaar)
        print("Address:", self.address)

    def transfer(self):
        print("Your current balance: ", self.bal)
        acc_no = int(input('Enter the beneficiary acc_no:'))
        ifsc_code = input("Enter the Beneficiary IFSC:")
        if acc_no in bank.cust_details and ifsc_code == bank.cust_details[acc_no].ifsc:
            amt = int(input("Enter the Transfer Amount:"))
            if self.bal >= amt:
                other = bank.cust_details[acc_no]
                other.bal += amt
                self.bal -= amt
                bank.concat_cust_tran(self.account_no, {'date': datetime.datetime.now(), 'type': 'Transfer(w)', 'amount': amt, 'bal': self.bal})
                bank.concat_cust_tran(other.account_no, {'date': datetime.datetime.now(), 'type': 'Transfer(w)', 'amount': amt, 'bal': other.bal})
                self.success()
                print("After transfer current balance", self.bal)
                return
            else:
                self.failed()
                print("Insufficient funds")
        else:
            print("Invalid account number or IFSC code")
        return

    @classmethod
    def home(cls, self):
        n = int(input("Choose the option:\n0.Account Details\n1.Check Bal\n2.Deposit\n3.Withdraw\n4.Mini Statement\n5.Update\n6.Pin Change\n7.Check ROI\n8.Transfer\n9.Exit"))
        print('*' * 150)
        match n:
            case 0:
                self.account_details()
            case 1:
                self.check_bal()
            case 2:
                self.deposit()
            case 3:
                self.withdraw()
            case 4:
                self.mini_statement()
            case 5:
                self.cust_update()
            case 6:
                self.pin_change()
            case 7:
                print("Today Rate of Interest", cls.roi)
            case 8:
                self.transfer()
            case _:
                return
        return cls.home(self)

    @staticmethod
    def success():
        print("|----------------------|")
        print("|Transaction is Success|")
        print("|----------------------|")

    @staticmethod
    def failed():
        print("|---------------------|")
        print("|Transaction is Failed|")
        print("|---------------------|")

obj = bank('krishna', 26, 'krishna@gmail.com', 9911776655, 'hyd', 987456321451,"DIVI0342", '1997-08-09', 'Male', 1234, 500)
#obj1 = bank('radha', 26, 'radha@gmail.com', 9911776655, 'hyd', '987456321451', 'abcd826g1', 'abcd82661', '1997-08-09', 'Male', 9876)
#obj=bank("NIKKI",21,"nikki@gmail.com",8075000354,"YARAM  NAGAR",807500035314,"DIVI0342","19-11-2023","male",5413,1500)
#obj1=bank("Balram",21,"balram28@gmail.com",8075001354,"KHAMMAM",745645879234,"DIVI0342","19-11-2023","male",2003,2000)

def main():
    acc_no = int(input("Enter the Account No:"))
    pin = int(input('Enter the pin no:'))
    if acc_no in bank.cust_details and bank.cust_details[acc_no].pin == pin:
        user = bank.cust_details[acc_no]
        bank.home(user)
    else:
        print("Invalid account no or pin")


main()
```
